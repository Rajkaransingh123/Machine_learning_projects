Test model on handwritten image 
-----------------------------------------------------------------------------------------------------------------------------

```python
#load model
from tensorflow.keras.models import load_model
model = load_model('handwritten_numeric_digits_classifier.h5')
```

```python
model.summary()
```

```python
model.input_shape
```

Load image
-----------------------------------------------------------------------------------------------------------------------------

```python
from PIL import Image
import numpy as np
```

```python
image = Image.open("C:\\Users\\rajka\\OneDrive\\Pictures\\captured images\\handwritten_digit_6.jpg")
```

```python
#convert the immage ot gray scale
gray_image= image.convert('L')
```

```python
#resize the image according to the  model
new_image =gray_image.resize((28,28))
```

```python
#convert to pixels
img = np.array(new_image)
```

```python
#check the max pixel value of  image
img.max()
```

```python
#Normalize the image
img = img/218
```

```python
#reshape according to the model 
final_img = img.reshape(1,28,28)
```

```python
#check if image is ready to input in the model
final_img.shape
```

```python
y_prob= model.predict(final_img)
```

```python
y_out = y_prob.argmax(axis=1)
y_out
```
